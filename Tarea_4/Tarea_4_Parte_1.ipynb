    * 2.1. Al instanciar un objeto, lee el archivo de datos y lo coloca en un dataframe interno en la clase.
    * 2.2. Cada vez que se le pidan datos, debe regresar los dos siguientes datos.
    * 2.3. Entregue un error si los datos se han agotado


```python
import pandas as pd

class datosIterable:
    def __init__(self, datos):
        self.df = pd.read_csv(datos, delimiter=",")
        ##print(self.df)
        self.index = 0
    
    def getDatos(self):
        
        if self.index+1 < len(self.df):
            lista = self.df.iloc[self.index:self.index+2]
            self.index += 2
        else:
            raise Exception("Los datos se han agotado.")
        return lista
            
data = datosIterable('iris.csv')
```

```python
##Prueba de la clase
while(True):
    print(data.getDatos())
```
